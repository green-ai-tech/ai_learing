```python
#1. 加载器
from langchain_community.document_loaders import TextLoader,PyPDFLoader,WebBaseLoader,CSVLoader
#2. 文档切割器
from langchain_text_splitters import RecursiveCharacterTextSplitter

file_path = "/Users/logicye/Code/ai_learning/assets/pdf/04tool工具的使用.pdf"
loader = PyPDFLoader(file_path)
docs =loader.load()
```

## 7.2 检索

```python
help(RecursiveCharacterTextSplitter)
```

```python
#切割

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 200,
    chunk_overlap = 20,
    length_function = len,       #python 的运算符len
    add_start_index = True,
)

chunks = text_splitter.split_documents(docs)
```

```python
type(chunks)
```

```python
type(chunks[0])
```

```python
len(chunks)
```

### 3）词嵌入
    - 特征提取（算法->模型【算子】：算法，算力，算子）
    - 模型：qwen-embeding3.5


```python
from langchain.embeddings import init_embeddings
from langchain.embeddings import Embeddings         #抽象类

emb = init_embeddings(model= "ollama:qwen3-embedding:0.6b")
```

```python
type(emb)
```

```python
vec = emb.embed_documents(chunks[0].page_content)
print(len(vec))
print(len(vec[0]))          #词潜入模型确定
```

### （4） 使用向量数据库

```python
from langchain_chroma import Chroma

#创建
db = Chroma(
    embedding_function =emb,
    persist_directory  ="/Users/logicye/Code/ai_learning/notebooks/04_大模型agent/vdb"
)

#插入数据
ids = db.add_documents(chunks)

ids[0]
#删除 修改


#查询（cos 相似度）
```

```python
# 查询 - 相似度查询
results = db.similarity_search("工具的使用",k=2)
print(type(results))
print(results[0])
```

```python
results = db.similarity_search_with_score("中间件调用",k = 2)
print(f"{results=}")
```

```python
results = db.max_marginal_relevance_search(
    "装饰器",
    k=3,
    fetch_k=5,
    lambda_mult=0.5     #0表示 最大的多样性； 1 表示最大的多样性
    )
print(results)
```
