```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model="ollama:qwen3.5:9b",
    #model_provider="ollama"

    #与供应商
    temperature=0.7,
    reasoning=True,
    num_predict=512,
    seed=40,
    format="json",
    keep_alive=True
    
)

print(type(model))
print(model)
```

- 模型的类型结构：
    - ChatOllama: 核心负责：
        - BaseChatModel
            - BaseLanguageModel
                - Runnable: invoke (单任务)，stream(流)，#batch ，#generate

- init_chat_model 函数是工厂函数：
    - 用来创建对象：
        该函数参数分为2:
            - 显示参数：与函数有关
            - 隐式参数：完全由对象构造器决定

OllamaLLM 与 ChatOllama:
  - OllamaLLM 是一个通用的语言模型接口，适用于各种语言模型提供商（如 OpenAI、Hugging Face、Ollama 等）。它提供了统一的方法来调用不同的语言模型。
  - ChatOllama 是 OllamaLLM 的一个具体实现，专门用于与 Ollama 服务进行交互。

- 没有写明供应商，自动腿导供应商


            - `gpt-...` | `o1...` | `o3...`       -> `openai`
            - `claude...`                         -> `anthropic`
            - `amazon...`                         -> `bedrock`
            - `gemini...`                         -> `google_vertexai`
            - `command...`                        -> `cohere`
            - `accounts/fireworks...`             -> `fireworks`
            - `mistral...`                        -> `mistralai`
            - `deepseek...`                       -> `deepseek`
            - `grok...`                           -> `xai`
            - `sonar...`                          -> `perplexity`
            - `solar...`                          -> `upstage`

- 本地模型与云上模型的差异：
    - 本地模型 控制模型更多的参数（可以二次开发为云服务）
    - 云上模型服务：
        - model
        - api_key
        - temperature
        - max_tokens：输出最大的token数
        - timeout:连接延时
        - num_retries:连接重试的次数

# 2. configurable_fields 与 config_prefix参数的使用

```python
from langchain.chat_models import init_chat_model

#构造一个动态模型
model = init_chat_model()         

#调用动态模型
response = model.invoke(
    input="n你是谁",
    config={    
        "configurable":{
            "model":"ollama:qwen3.5:9b",
            "temperature":0.7,
            "num_predict":512,
            "base_url":"http://localhost:11434",
        }
    }
)
print(response.content)
```

# 3. stream的使用

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("ollama:qwen3.5:9b")

response = model.stream("用java 写一个冒泡排序算法。")

print(type(response))
```

```python
for msg in response:
    # print(type(msg))
    # break
    # print(msg.content)
    print(msg.content, end="", flush=True)
print()
```
