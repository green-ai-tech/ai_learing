```python
from transformers import pipeline
import torch
import warnings
warnings.filterwarnings("ignore")

# 修正后的配置
pipe = pipeline(
    task="audio-classification",
    model="/Users/logicye/Code/models/mms-lid-126",  # 模型路径
    device="mps",                                    # 苹果 GPU
    torch_dtype=torch.float32                        # 先使用 float32 而不是 bfloat16
)

# 测试识别
result = pipe("0a196374_nohash_0.wav", top_k=10)

print("\n识别结果（语种 + 置信度）：")
for res in result:
    print(f"{res['label']:>20} | {res['score']:.4f}")
```
