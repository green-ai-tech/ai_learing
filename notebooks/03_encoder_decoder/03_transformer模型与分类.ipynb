```python
from transformers import Qwen2TokenizerFast, Qwen3ForCausalLM

path = "/Volumes/AI/models/local/Qwen3-4B"

model = Qwen3ForCausalLM.from_pretrained(path)
tokenizer = Qwen2TokenizerFast.from_pretrained(path)

print("✅ 模型加载成功！")
```

```python
# 1. 预处理
msg = [{
        "role":"user",
        "content":"你好！"
    }]

inputs = tokenizer.apply_chat_template(
    msg,
    add_generation_prompt= True,
    tokenize = True,
    return_tensors = "pt"
)
print(input)
```

```python
#2. 推理

# 如果要获取 logits,直接调用模型
outputs = model(**inputs)
print(outputs)
print(outputs.logits.shape)  
print("="*80)

# 如果要生成文本,使用 model.generate()
generated = model.generate(**inputs)
print(generated)  # 这包含生成的 token ID
```
