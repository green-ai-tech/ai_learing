# 1. 实现编码器与解码器

```python
import torch
class AutoEncoderAndDecoder(torch.nn.Module):
    def __init__(self):
        super(AutoEncoderAndDecoder,self).__init__()
        # 定义网络层（分组）
        self.encoder = torch.nn.Sequential(
            torch.nn.Linear(28*28, 256), # 一层（运算，激活，dropout，batchnorm） 
            torch.nn.ReLU(inplace=True),
            torch.nn.Dropout(0.0, inplace=True),
            
            torch.nn.Linear(256, 64), # 
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(64, 16), # 
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(16, 3), # 

            

        )
        self.decoder = torch.nn.Sequential(
            torch.nn.Linear(3, 16),
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(16, 64),
            torch.nn.ReLU(inplace=True),           
            torch.nn.Linear(64, 256),
            torch.nn.ReLU(inplace=True),           
            torch.nn.Linear(256, 28* 28),
           
        )
            
    def forward(self, x):
        y_enc = self.encoder(x)
        y_dec = self.decoder(y_enc)
        return y_enc, y_dec
        
```

# 2. 训练编码器与解码器

- 使用手写字符数据集：
    - 解码器生成手写字符
    - 

```python
import torch
import torchvision
transform = torchvision.transforms.Compose(
    [
        torchvision.transforms.ToTensor(),
    ]
) 
ds_train = torchvision.datasets.MNIST(
    root="./ds", 
    download=True, 
    train=True, 
    transform=transform
)
ld_train = torch.utils.data.DataLoader(
    ds_train, 
    shuffle=True, 
    batch_size = 100
)
model = AutoEncoderAndDecoder()

# for x, _ in ld_train:
#     encode,decode = model(x.view(x.shape[0], -1))
#     print(encode)
#     print(decode)
#     decode = decode.view(-1, 1, 28, 28)
#     torchvision.utils.save_image(decode, "1.jpg")
#     break

# 训练：模型，数据集，优化器，损失函数，超参数
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

f_loss = torch.nn.MSELoss(reduction="mean")

epoches = 100

# 设置模型的保存目录
outputs_dir = "outputs/"
import os
if not os.path.exists(outputs_dir):
    os.mkdir(outputs_dir)
```

```python
# 开始训练
device = "cuda" if torch.cuda.is_available() else "cpu"

model.to(device)
for e in range(epoches):

    # ========================================================
    def closure():
        global v_loss, x, y_                 # 设置为全局变量
        # 梯度清零
        optimizer.zero_grad()
        # 预测
        x = x.view(x.shape[0], -1)
        _, y_ = model(x)
        # 计算损失
        v_loss = f_loss(y_, x)
        # 自动求导
        v_loss.backward()
        # 梯度更新（不做）
        return v_loss             # 其中包含梯度
    # ========================================================
    for x, _ in ld_train:
        x = x.to(device)
        optimizer.step(closure)  # 梯度更新 step()，step（训练回调函数）
    # 输出损失值，保存输出
    print(F"损失值：{v_loss:.4f}")
    torch.save(model.state_dict(),outputs_dir + "/model.pt")
    if e % 20 == 0:
        
        torchvision.utils.save_image(
            y_.view(y_.shape[0], 1, 28, 28), # N C H W
            outputs_dir + f"out_{e:04d}.jpg",
        )
```
