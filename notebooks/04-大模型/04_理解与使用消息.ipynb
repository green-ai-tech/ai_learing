````python
# 导入必要的模块
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.messages import AIMessage, HumanMessage

# 创建AI消息的提示模板
template_ai = AIMessagePromptTemplate.from_template("我是{role},负责解惑答疑。")
# 创建用户消息的提示模板
template_user = HumanMessagePromptTemplate.from_template("请回答：{question}")

# 创建聊天提示模板，包含系统消息、历史消息占位符和用户输入
template = ChatPromptTemplate.from_messages([
   ("system", "你是一个软件工程师"),                      # 系统角色定义
   MessagesPlaceholder(variable_name="chat_history"),  # 用于存放历史对话的占位符
   ("user", "{inputs}")                                # 用户输入
])

# 模拟一个历史对话记录
chat_history = [
    HumanMessage(content="杭州天气如何"),   # 用户提问杭州天气
    AIMessage(content="抱歉，臣妾做不到；")  # AI回复无法回答
]

# 使用模板生成输入消息，包含历史对话和新的用户输入
input_message = template.invoke({
    "chat_history": chat_history,  # 注入历史对话
    "inputs": "明天出行，有何建议"    # 新的用户问题
})

# 打印生成的提示消息
print(input_message)
````

- 形成提示词与模版之间的构建管道

````python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.chat_models import init_chat_model

#创建提示词模版
template = ChatPromptTemplate.from_messages([
    ("system", "你是一个{role},提供简洁精准地回答，直接给出答案"), 
    ("user", "{inputs}")                                
])

#创建模型
model = init_chat_model("ollama:qwen3.5:0.8b")

#创建模型
chain = template | model


#调用
val = chain.invoke({"role":"软件工程师","inputs":"python如何实现冒泡排序？"})
print(val)
print(val.content)
````
