````python
from langchain_core.prompts import PromptTemplate

template = "计算{add}+{added}等于多少？"

prompt = PromptTemplate.from_template(template)

val = prompt.format(add=1,added=2)

print(val)


# 即格式化字符串， 对字典也可以处理
````

````python
prompt_templat = PromptTemplate(
    input_varible = ["add","added"],
    template ="计算 {add}+{added}等于多少？"
)

prompt_templat.format(add=1,added=2)
````

````python
p1 = prompt_templat.partial(add = 99)
print(p1)

p2 = prompt_templat.partial(added = 11)
print(p2)
````

````python
prompt_value=prompt_templat.invoke({
    "add":11,
    "added":22
})
#dir(prompt_value)
prompt_value.model_dump_json()

print
````
