````python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.messages import AIMessage, HumanMessage, SystemMessage

template = ChatPromptTemplate.from_messages([
    ("system", "你是一个大模型工程师"),                       # 系统信息
    MessagesPlaceholder(variable_name="chat_history"),      # 历史聊天（建议保留最后n条） 
    ("user", "{inputs}"),                                   # 问题 
])

# 模拟一个历史对话
chat_history = [
    HumanMessage(content="杭州天气怎样"),
    AIMessage(content="我不能提供今天的天气，您可以通过网站查询。")
]

input_message = template.invoke({
    "chat_history": chat_history,
    "inputs": "明天出现，给我一个准备的建议。"
})
print(input_message)
````

- 形成提示词与模型之间构建管道。

````python
from langchain_core.prompts import ChatPromptTemplate
from langchain.chat_models import init_chat_model

# 创建提示词模版
template = ChatPromptTemplate.from_messages([
    ("system", "你是{role}, 提供简洁精准的回答，直接给出答案。"),
    ("user", "{inputs}"),
])
# 创建模型
model = init_chat_model("ollama:qwen3.5:0.8b")
# 创建管道(链chain)
chain = template | model 
# 调用
val = chain.invoke({
    "role": "计算专家",
    "inputs":"45的平方根是多少？"
})
val
````
