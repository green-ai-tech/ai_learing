- 图像阈值处理
- 图像轮廓抽取
- 图像轮廓绘制
- 图像轮廓面积

```python
import cv2
import matplotlib.pyplot as plt

img_path = "imgs/scene.jpg"
#读取文件
mat_img = cv2.imread(img_path, cv2.IMREAD_COLOR_RGB)
mat_img2 = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)

#自适应分割（图像预处理）
dst = cv2.adaptiveThreshold(mat_img2, 10, cv2.BORDER_REPLICATE, cv2.THRESH_BINARY_INV, 3, 10)
#提取轮廓
contours,heridency = cv2.findContours(dst, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
#标记轮廓
cv2.drawContours(mat_img, contours, -1, (255,0,255), 3)

#计算轮廓面积
area = 0
for i in contours:
    area += cv2.contourArea(i)
print(area)

#图像show
plt.subplot(121)
plt.imshow(dst + 50, cmap="gray")
plt.axis("off")

plt.subplot(122)
plt.imshow(mat_img)

plt.show()
```
