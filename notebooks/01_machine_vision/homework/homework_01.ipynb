```python
import cv2
import time
import numpy as np

# 获取当前时间信息
now = time.localtime()
year = now.tm_year
month = now.tm_mon
day = now.tm_mday
hour = now.tm_hour
minute = now.tm_min
second = now.tm_sec

#获取摄像头视频
dev = cv2.VideoCapture(0)
fps = int(dev.get(cv2.CAP_PROP_FPS))
frames = int(dev.get(cv2.CAP_PROP_FRAME_COUNT))
width = int(dev.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(dev.get(cv2.CAP_PROP_FRAME_HEIGHT))


while True:
    status, frame = dev.read()
    if status:
        #----------------------图像处理-----------------------


        #绘制文本信息如： 2028/04/12--08:00:00,FPS=30
        cv2.putText(
            frame,
            text=f"{year}/{month}/{day}--{hour}:{minute}:{second},FPS={fps}",
            org=(0,height-5),
            fontFace=0,
            fontScale=1,
            color= (0,255,255),
            thickness=2) 

        #给视频周围加上边框
        cv2.rectangle(frame,(0, 0,width,height),(255,255,255),5)


        #绘制了两个左右对称的椭圆
        cv2.ellipse(frame, 
                    center=(200,200), 
                    axes=(100,60), 
                    angle=0, 
                    startAngle=0, 
                    endAngle=360, 
                    color=(255,0,255), 
                    thickness=2, 
                    lineType=None, 
                    shift=None)
        
        cv2.ellipse(frame, 
                    center=(width-200,200), 
                    axes=(100,60), 
                    angle=0, 
                    startAngle=0, 
                    endAngle=360, 
                    color=(255,0,255), 
                    thickness=2, 
                    lineType=None, 
                    shift=None)

        #趣味玩法,加上蒙版效果
        frame=cv2.Sobel(frame,-1,dx=0,dy=1,ksize=3,scale=1.0,delta=180)
        
        #----------------------------------------------------
        cv2.imshow("Image", frame)
        # 按键退出
        if cv2.waitKey(80) == ord('q'):
            break
    else:
        print("读取失败！")
        break
# 完成任务释放资源
dev.release()
cv2.destroyAllWindows()
```
